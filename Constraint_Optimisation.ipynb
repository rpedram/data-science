## Problem
A project involves three activities a1, a2 and a3. a3
can start only after a1 and a2 are completed.
The amount of work done by a single worker in a
day for the activities are 2, 3 and 1 units
respectively. The total amount of work for the
activities are 4, 3 and 30 units respectively. The
total number of workers available is 15. How
many workers should be assigned to activity 1
such that the total time taken for the project is a
minimum?



````python
def calculate_total_time(workers_a1):
    # Calculate the time taken for each activity based on the number of workers assigned
    time_a1 = 4 / (workers_a1 * 2)

    # Handle the case where workers_a1 is 15 to avoid division by zero
    if workers_a1 == 15:
        time_a2 = float('inf')  # Set time_a2 to infinity if no workers are assigned
    else:
        time_a2 = 3 / ((15 - workers_a1) * 3)

    time_a3 = 30 / (15 * 1)

    # Activity a3 can start only after a1 and a2 are completed
    total_time = max(time_a1, time_a2) + time_a3

    return total_time

def find_optimal_workers():
    min_time = float('inf')
    optimal_workers_a1 = 0

    # Iterate through all possible assignments of workers to activity a1
    for workers_a1 in range(1, 16):  # Workers can be from 1 to 15
        total_time = calculate_total_time(workers_a1)

        # Update the minimum time and optimal number of workers if a better solution is found
        if total_time < min_time:
            min_time = total_time
            optimal_workers_a1 = workers_a1

    return optimal_workers_a1, min_time

# Find the optimal number of workers for activity a1
optimal_workers_a1, min_time = find_optimal_workers()

print(f"Optimal number of workers for activity a1: {optimal_workers_a1}")
print(f"Minimum total time for the project: {min_time:.2f} days")
````

# Problem 2
A project involves four activities A1, A2, A3 and A4. Characteristics of these activities
are as follows:
• A1: This activity requires 50 units of work to be completed. A worker can
complete 2 units of work in a day.
• A2: This activity requires 75 units of work to be completed. A worker can
complete 3 units of work in a day.
• A3: This activity requires 60 units of work to be completed. A worker can
complete 4 units of work in a day.
• A4: This activity requires 11 units of work to be completed. A worker can
complete 1 unit of work in a day.
A1, A2 and A3 can be scheduled in parallel or in any sequence. However, activity A4
can be started only after the activities A1, A2 and A3 are completed. No activity
can be completed within a fraction of a day, that is, workers doing an activity will
be released only at the end of the day and not in between. All the activities have to
be completed within 7 days. A worker is paid Rs. 500 per day for all the days for
the entire duration of the project, including any days the worker is idle. Work out
the schedule such that the labour cost is the minimum.

## Code P2

This code uses functions to compute optimisation. We can also run it in a loop. But this is more elegant and easier to troubleshoot. Select the number of days you want to get the task completed in the form.

````python
# @title Max day {"run":"auto","form-width":"250px"}
day_constraint = 7 # @param {"type":"integer"}
import math

def calculate_days(work_units, work_rate, workers):
    return math.ceil(work_units / (work_rate * workers))

def calculate_total_cost(workers_a1, workers_a2, workers_a3, workers_a4):
    days_a1 = calculate_days(50, 2, workers_a1)
    days_a2 = calculate_days(75, 3, workers_a2)
    days_a3 = calculate_days(60, 4, workers_a3)
    days_a4 = calculate_days(11, 1, workers_a4)

    max_days = max(days_a1, days_a2, days_a3)
    total_days = max_days + days_a4

    if total_days > day_constraint:
        return float('inf'), total_days  # If the total days exceed the constraint, the cost is infinite

    total_workers = workers_a1 + workers_a2 + workers_a3 + workers_a4
    total_cost = total_workers * 500 * total_days

    return total_cost, total_days

def find_optimal_schedule():
    min_cost = float('inf')
    optimal_workers = (0, 0, 0, 0)
    optimal_total_days = 0

    # Maximum reasonable number of workers for each activity
    max_workers = 25

    for workers_a1 in range(1, max_workers + 1):
        for workers_a2 in range(1, max_workers + 1):
            for workers_a3 in range(1, max_workers + 1):
                for workers_a4 in range(1, 12):  # A4 can have up to 11 workers
                    total_cost, total_days = calculate_total_cost(workers_a1, workers_a2, workers_a3, workers_a4)

                    if total_cost < min_cost:
                        min_cost = total_cost
                        optimal_workers = (workers_a1, workers_a2, workers_a3, workers_a4)
                        optimal_total_days = total_days

    return optimal_workers, min_cost, optimal_total_days

# Find the optimal schedule
optimal_workers, min_cost, optimal_total_days = find_optimal_schedule()

if min_cost == float('inf'):
    print(f'Not feasible to complete the task within the schedule of {day_constraint} days')
else:
    print(f"Optimal number of workers for A1: {optimal_workers[0]}")
    print(f"Optimal number of workers for A2: {optimal_workers[1]}")
    print(f"Optimal number of workers for A3: {optimal_workers[2]}")
    print(f"Optimal number of workers for A4: {optimal_workers[3]}")
    print(f"Minimum total labor cost: Rs. {min_cost}")
    print(f"Total number of days: {optimal_total_days}")
````



### Explanation:

   - The `math.ceil` function is used to round up the number of days to the nearest whole number, ensuring that days are not split.

2. **calculate_total_cost(workers_a1, workers_a2, workers_a3, workers_a4)**:
   - This function calculates the total cost of labor for the given number of workers for each activity.
   - It first calculates the number of days required to complete each activity using the `calculate_days` function.
   - It then determines the maximum number of days required to complete activities A1, A2, and A3.

   - If the total days exceed the day constraint, the cost is set to infinity.
   - Otherwise, the total cost is calculated as the product of the total number of workers, the daily wage (Rs. 500), and the total number of days.
   - The function returns both the total cost and the total number of days.

3. **find_optimal_schedule()**:
   - This function iterates through all possible assignments of workers to activities A1, A2, and A3 (from 1 to a maximum reasonable number of workers).

````python
from tabulate import tabulate

def optimized_work_schedule():
    # Define activities and their properties
    activities = {
        "A1": {"work_units": 100, "work_rate": 2},
        "A2": {"work_units": 75, "work_rate": 3},
        "A3": {"work_units": 60, "work_rate": 4},
        "A4": {"work_units": 35, "work_rate": 1}
    }

    max_workers = 25
    max_days = 8
    best_schedule = None
    best_cost = float('inf')  # Initialize with a very high cost
    best_labor_employed = max_workers  # Initialize with max_workers

    # Iterate backwards from the maximum number of workers to find the optimal solution
    for num_workers in range(max_workers, 0, -1):
        # Initialize the schedule
        schedule = {day: {"A1": 0, "A2": 0, "A3": 0, "A4": 0} for day in range(1, max_days + 1)}

        # Assign workers to A1, A2, and A3, ensuring we don't exceed the total number of workers
        remaining_work = {activity: activities[activity]["work_units"] for activity in ["A1", "A2", "A3"]}
        for day in range(1, max_days + 1):
            available_workers = num_workers
            for activity in ["A1", "A2", "A3"]:
                if remaining_work[activity] > 0 and available_workers > 0:
                    # Ensure we don't assign more workers than available or needed
                    workers_assigned = min(available_workers, remaining_work[activity] // activities[activity]["work_rate"])
                    schedule[day][activity] = workers_assigned
                    work_done = workers_assigned * activities[activity]["work_rate"]
                    remaining_work[activity] -= work_done
                    available_workers -= workers_assigned

        # Check if all activities A1, A2, and A3 are completed
        all_completed = all(remaining_work[activity] == 0 for activity in ["A1", "A2", "A3"])


        if all_completed:
            # Find the latest completion day among A1, A2, and A3
            completion_day = max(
                next((day for day, assignments in schedule.items() if assignments.get("A1", 0) == 0), 1),
                next((day for day, assignments in schedule.items() if assignments.get("A2", 0) == 0), 1),
                next((day for day, assignments in schedule.items() if assignments.get("A3", 0) == 0), 1)
            )


            # Check if A4 can be completed within the remaining days (including the completion day)
            remaining_days = max_days - completion_day + 1
            if remaining_days * num_workers >= activities["A4"]["work_units"]:

                # Assign all workers to A4 only after A1, A2, and A3 are completed on each day
                remaining_work_A4 = activities["A4"]["work_units"]

                for day in range(completion_day, max_days + 1):
                    if remaining_work_A4 > 0 and all(schedule[day][activity] == 0 for activity in ["A1", "A2", "A3"]):
                        # Assign only the number of workers needed to complete A4, or all workers if there's enough work
                        workers_assigned_to_A4 = min(num_workers, remaining_work_A4)
                        schedule[day]["A4"] = workers_assigned_to_A4
                        remaining_work_A4 -= workers_assigned_to_A4 * activities["A4"]["work_rate"]

                # Ensure A4 is fully completed
                if remaining_work_A4 == 0:
                    # Find the actual completion day (the last day where A4 has workers assigned)
                    completion_day = next(
                        (day for day, assignments in reversed(schedule.items()) if assignments["A4"] > 0),
                        None  # Default to None if A4 was never started
                    )

                    if completion_day:
                        # Calculate labor cost based on actual completion day
                        labor_cost = 500 * num_workers * completion_day

                        # Calculate the maximum labor employed on any single day
                        max_labor_employed = max(sum(assignments.values()) for assignments in schedule.values())

                        # Check if this is a better solution (lower cost)
                        if labor_cost < best_cost or (labor_cost == best_cost and max_labor_employed < best_labor_employed):
                            best_schedule = schedule
                            best_cost = labor_cost
                            best_labor_employed = max_labor_employed
                            best_completion_day = completion_day # Store the best completion day

    if best_schedule:
        # Prepare data for tabulate, only up to best_completion_day
        table_data = [["Day", "A1", "A2", "A3", "A4"]]
        for day, assignments in best_schedule.items():
            if day <= best_completion_day:
                table_data.append([day, assignments['A1'], assignments['A2'], assignments['A3'], assignments['A4']])

        # Print the schedule and cost
        print(tabulate(table_data, headers="firstrow", tablefmt="grid"))
        print(f"\nTotal Labor Cost: Rs. {best_cost}")
        print(f"Total Labor Employed: {best_labor_employed}")

        # Print the minimum days to complete the task (the actual completion day)
        print(f"Minimum days to complete the task: {best_completion_day}")
    else:
        print("No feasible solution found within the given constraints.")

# Run the optimization
optimized_work_schedule()
````

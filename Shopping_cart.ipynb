```python
import uuid # This is a library for session ID creation
# Define product categories
import random # Generate product ID new products
categories = ["Liquor", "Coats", "Jackets", "Caps"]

# Define sample products with category IDs and prices
products = [
    {
        "product_id": 'a',
        "product_name": "Beer",
        "category_id": 0,
        "price": 1000,
        "stock":150,
    },
    {
        "product_id": 'b',
        "product_name": "Winter Coat",
        "category_id": 1,
        "price": 2000,
        "stock":100,
    },
    {
        "product_id": 'c',
        "product_name": "Denim Jacket",
        "category_id": 2,
        "price": 1500,
        "stock":100,
    },
    {
        "product_id": 'd',
        "product_name": "Beanie",
        "category_id": 3,
        "price": 500,
        "stock":100,
    },
]
# Store Login credentials in a dictonary
password_data ={'admin':'admin','user1':'me'}
#Declare admin session ID
admin_session_id = str()
# Store user cart items and quantities
user_cart = {}
user_sessionid=str() #Declare user session ID

# Define payment options
payment_options = ["Net banking", "PayPal", "UPI"]

# User login and admin login functions
def user_login():
    # Check for valid user credentials
    while True:

        username = input("Enter your username(default user1): ")
        password = input("Enter your password(default me): ")

        if username in  password_data:
            if password == password_data[username]:
                break

        print('Wrong credentails reneter username passward')


    # Store session for user
    global usercart_sessionid
    usercart_sessionid = generate_session_id()

    # Add session ID to user cart
    #user_cart['session_id'] = session_id

    # Welcome message for logged-in user
    print("Welcome, " + username + "! Your session ID is " + usercart_sessionid)
    input('Press key yo go further')

    # Display product catalog
    display_product_catalog()
    input('read catlog and press key to go further')

    # Manage cart
    manage_cart(usercart_sessionid)

def admin_login():
    # Check for valid admin credentials
    while True:

        username = input("Enter your username(default admin): ")
        password = input("Enter your password(default admin): ")

        if username in  password_data:
            if password == password_data[username]:
                break

        print('Wrong credentails reneter username passward')

    # Store session for admin
    global admin_session_id
    admin_session_id = generate_session_id()

    # Add session ID to admin cart
    #user_cart['session_id'] = session_id

    # Welcome message for logged-in admin
    print("Welcome,", str(username), 'Your session ID is  ' , admin_session_id)

    # Display product catalog
    display_product_catalog()

    # Manage cart (admin has no cart)
    manage_cart(admin_session_id)

    # Admin functions: add, modify, and remove products and categories
    admin_functions(admin_session_id)

# Function to display product catalog
def display_product_catalog():
    print("\nProduct Catalog")
    print("----------------")
    for product in products:
        print("Product ID: " + str(product['product_id']))
        print("Product Name: " + product['product_name'])
        print("Category: " + categories[product['category_id']])
        print("Price: " + str(product['price']) + "\n")
        print("Stock Held is: " + str(product['stock']) + "\n")

# Function to manage cart
def manage_cart(session_id):
    global admin_session_id

    while True:

        # Display cart contents
        print("\nCart Contents")
        print("----------------")
        #This prints the cort contents for items in user cart
        #less the first item which is session ID
        for product_id, quantity in user_cart.items():
            ls=[product['product_name'] for product in products if product['product_id']== product_id]
            print(f'Item ID {product_id}  Name :{ls}  {str(quantity)}')


        #print('sesion ID',session_id,' admin session id is  ',admin_session_id) debugging ssession id variable
# Reject Admin Users to go further
        if  session_id == admin_session_id:
            print(' You are admin user: cant proceed amend cart \n')
            print('Pres any key to continue to admin menu')
            input()
            break

        # Add or remove items from cart
        choice = input("\nWhat do you want to do?\n1. Add Item\n2. Remove Item\n3. Checkout\n0. Exit\nEnter your choice: ")

        if choice == '1':
            add_item_to_cart(session_id)
        elif choice == '2':
            remove_item_from_cart(session_id)
        elif choice == '3':
            proceed_to_checkout(session_id)
        elif choice == '0':
            break
        else:
            print("Invalid choice. Please enter a valid option.")

# Function to add item to cart
def add_item_to_cart(session_id):
    #Show ID on input
    print([product['product_id'] for product in products])
    # Check if product ID is valid
    product_id = input("Enter product ID: ")
    #trouble shoot
    print([product['product_id'] for product in products])
    if product_id not in [product['product_id'] for product in products]:
        print("Invalid product ID. Please enter a valid product ID.")
        return

    # Check if product quantity is valid
    quantity = int(input("Enter quantity: "))
    if quantity < 1:
        print("Quantity should be 1 or greater. Please enter a valid quantity.")
        return

    # Check if product already exists in cart
    if product_id in user_cart:

        existing_quantity = user_cart.get(product_id)
        print(f'existing cart quantity is {existing_quantity}')
        cart_stock=0
        for product in products:
            if product['product_id']==product_id:
               cart_stock=product['stock']
        print(f'cart stock quantity is {cart_stock}')
        input('stop check 163 debug')


        #existing_quantity = user_cart[product_id][quantity]
        updated_quantity = existing_quantity + quantity
        print(f'updated quantity is {updated_quantity}')
        # Horrible way!! of workking with dictionary. Compamise between array and dics
        if updated_quantity > cart_stock:
            print("Not enough stock for the specified quantity. "
                  "Please enter a quantity within the available stock.")
            return
        else:
            user_cart[product_id]=updated_quantity
            print("Product added to cart successfully.")

        return

    # Add product to cart if it doesn't exist
    user_cart[product_id]=quantity

    #user_cart[product_id] = quantity
    print("Product added to cart successfully.")

# Function to remove item from cart
def remove_item_from_cart(session_id):
    # Check if product ID is valid
    print(user_cart)
    product_id = input("Enter product ID: ")
    if product_id not in user_cart:
        print("Product not found in cart. Please enter a valid product ID.")
        return

    # Remove product from cart
    del user_cart[product_id]
    #user_cart.pop(product_id)
    print("Product removed from cart successfully.")
    return

# Function to proceed to checkout
def proceed_to_checkout(session_id):
    global user_cart
    # Display checkout options
    for index, option in enumerate(payment_options):
        print(str(index + 1) + ". " + option)

    # Choose payment method
    payment_option = int(input("Enter your preferred payment option: "))

    # Simulate checkout process based on selected option
    if payment_option == 1:
        print("Net banking checkout in progress...")
        print("Your order has been placed successfully.")
    elif payment_option == 2:
        print("PayPal checkout in progress...")
        print("Your order has been placed successfully.")
    elif payment_option == 3:
        print("UPI checkout in progress...")
        print("Your order has been placed successfully.")
    else:
        print("Invalid payment option. Please select a valid option.")

    # Now that payment is made update the stock and empty cart go back to main
    #Update stock
    print(user_cart)
    for product_id in user_cart:

        for product in products:

            if product['product_id'] == product_id:
                print( 'product id in cart',product['product_id'])
                product['stock']=product['stock']-user_cart[product_id]
                print( 'Revised stock of ', product['product_id'], 'is ', product['stock'])

    # empty cart reset
    print('you have been logged off')
    user_cart={}
    #Go back to main screen
    main()

# Function to generate session ID
def generate_session_id():
    return str(uuid.uuid4())

# Function for admin functions: add, modify, and remove products and categories
def admin_functions(session_id):
    while True:
        print("\nAdmin Functions")
        print("----------------")
        print("1. Add Product")
        print("2. Modify Product")
        print("3. Remove Product")
        print("4. Add Category")
        print("5. Remove Category")
        print('6. Display Product catalouge')
        print("0. Exit Admin Mode")
        print("Enter your choice: ")

        choice = int(input())

        if choice == 1:
            # Add product
            add_product(session_id)
        elif choice == 2:
            # Modify product
            modify_product(session_id)
        elif choice == 3:
            # Remove product
            remove_product(session_id)
        elif choice == 4:
            # Add category
            add_category(session_id)
        elif choice == 5:
            # Remove category
            remove_category(session_id)
        elif choice ==6:
            #Display Catlogue
            display_product_catalog()
        elif choice == 0:
            break
        else:
            print("Invalid choice. Please enter a valid option.")
#user_login()
# Function to add product
def add_product(session_id):
    # Get product details from user
    product_name = input("Enter product name: ")
    product_price = int(input("Enter product price: "))
    #Print Categories
    for index,value in enumerate(categories):
        print('index ',index,' ',value,end=' ')
    product_category = int(input("Enter product category ID: "))
    product_stock = int(input("Enter product stock: "))

    # Check if product category exists
    #Category IDs starting from zero
    if product_category not in range(len(categories)):
        print("Invalid product category. Please enter a valid category ID.")
        return
    #Generate a randoms int for product ID
    rndstring=""
    for i in range(5):
        rndstring=rndstring + chr(random.randint(ord('a'),ord('z')))
    # Add product to list of products
    new_product = {
        "product_id": rndstring,
        "product_name": product_name,
        "category_id": product_category,
        "price": product_price,
        "stock": product_stock
    }
    products.append(new_product)
    print("Product added successfully.")

# Function to modify product
def modify_product(session_id):
    # Get product ID from user
    product_id = input("Enter product ID to modify: ")

    # Check if product exists
    if product_id not in [product['product_id'] for product in products]:
        print("Product with the given ID does not exist.")
        return

    # Get updated product details from user
    updated_product_name = input("Enter new product name (leave blank if you don't want to change): ").strip()
    updated_product_price = input("Enter new product price integer(leave blank if you don't want to change): ").strip()
    #Print Categories
    for index,value in enumerate(categories):
        print('index ',index,' ',value,end=' ')
    updated_product_category = input("Enter new product category ID INTEGER (leave blank if you don't want to change): ").strip()
    updated_product_stock = input("Enter new product stock INTEGER(leave blank if you don't want to change): ").strip()

    # Update product details
    for product in products:
        if product['product_id'] == product_id:
            if updated_product_name:
                product['product_name'] = updated_product_name
            if updated_product_price !="":
                product['price'] = int(updated_product_price)
            if updated_product_category !="":
                product['category_id'] = int(updated_product_category)
            if updated_product_stock !="":
                product['stock'] = int(updated_product_stock)
            print("Product modified successfully.")
            return

# Function to remove product
def remove_product(session_id):
    # Get product ID from user
    product_id = input("Enter product ID to remove: ")

    # Check if product exists
    if product_id not in [product['product_id'] for product in products]:
        print("Product with the given ID does not exist.")
        return

    # Remove product from the list of products
    for product in products:
        if product['product_id'] == product_id:
            products.remove(product)
            print("Product removed successfully.")
            return

# Function to add category
def add_category(session_id):
    #Print Categories
    for index,value in enumerate(categories):
        print('Existing cat',index, ',index ',value,end=' ')
    # Get new category name from user
    new_category = input("Enter new category name: ")

    # Add new category to the list of categories
    if new_category not in categories:
        categories.append(new_category)
        print("Category added successfully.")
    else:
        print("Category with the same name already exists.")

# Function to remove category
def remove_category(session_id):
    #Print Categories
    for index,value in enumerate(categories):
        print('Existing cat',index, ',index,' ,value,end=' ')
    # Get category name from user
    removed_category = input("Enter category name to remove: ")
    # Check if removed cat in the exisiting catlog
    list1= [product['category_id'] for product in products]
    catcheck=[]
    for i in range(len(list1)):
        catcheck.append(categories[list1[i]])
    if removed_category in catcheck:
        print('Category linked to existing product \n mod product first')
        return
    # Check if category exists
    if removed_category not in categories:
        print("Category with the given name does not exist.")
        return

    # Remove category from the list of categories
    categories.remove(removed_category)
    print("Category removed successfully.")

# Main function to start the shopping app

def main():
    while True:
        print("\nWelcome to the Demo Marketplace")

        # Display login options
        print("1. User Login")
        print("2. Admin Login")
        print("0. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            # User login
            user_login()
        elif choice == 2:
            # Admin login
            admin_login()
        elif choice == 0:
            print("Thank you for visiting the Demo Marketplace!")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

# Start the shopping app
if __name__ == "__main__":
    main()

```

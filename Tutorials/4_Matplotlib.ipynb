````python
import matplotlib.pyplot as plt
y=[1,4,6,18,20,36,54,72]
x1=[1,15,30,43,58,72,77,84]
x2=[1,8,14,32,44,68,71,80]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
line1=ax.plot(x1,y,'ys-') #Solid line with yellow color (y),square marker(s),-solid line
line2=ax.plot(x2,y,'go--')#dashed line with green color and circle marker
ax.legend(labels=('Iphone','Ipad'),loc='lower right')
ax.set_title("Friends referring iphone and ipad")
ax.set_xlabel("Item")
ax.set_ylabel('Sales')
plt.show()
````

````python
plt.xlabel("Range")
plt.ylabel("Various plots")
plt.title("More than one plot")
ex_plot3X=np.arange(0,math.pi*2,0.05)
ex_plot31Y=np.cos(ex_plot3X)
ex_plot32Y=np.sin(ex_plot3X)
ex_plot33Y=-np.cos(ex_plot3X)
plt.subplot(221)
plt.xlabel("Range")
plt.ylabel("Cosine")
plt.title("Cos plot")
plt.plot(ex_plot3X,ex_plot31Y, color='blue', linestyle='--',linewidth=1)
plt.subplot(222)
plt.xlabel("Range")
plt.ylabel("Sine")
plt.title("Sine plot")
plt.plot(ex_plot3X,ex_plot32Y, color='green', linestyle='--',linewidth=1)
plt.subplot(224)
plt.xlabel("Range")
plt.ylabel("- Cosine")
plt.title("- Cos plot")
plt.plot(ex_plot3X,ex_plot33Y, color='red', linestyle='--',linewidth=1)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.01)
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.1,0.1,0.2,0.2])
y=np.sin(x)
z=np.cos(x)
axes1.plot(x,y,'r')
axes2.plot(x,z,'b')
axes1.set_title('My Sine Plot')
axes2.set_title('My small cosine plot')
````

````python
fig,a=plt.subplots(4,3)
a[0][0].plot(x,y)
a[3][2].plot(x,z)
plt.show()
````

````python
#plt.subplot2grid(shape,location,rowspan,colspan)
import matplotlib.pyplot as plt
a1=plt.subplot2grid((3,3),(0,0),colspan=2)
a2=plt.subplot2grid((3,3),(0,2),rowspan=3)
a3=plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)
import numpy as np
x=np.arange(0,math.pi*2,0.01)
y=np.sin(x)
z=np.cos(x)
z1=-np.cos(x)
a1.plot(x,y)
a1.grid(True)
plt.xlabel("Hello")
a2.plot(x,z)
a2.grid(color='r',ls='--',lw=1)
a3.plot(x,z1)
plt.tight_layout()
plt.show()
````

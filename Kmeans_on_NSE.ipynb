## Distance in clustering
In clustering datapoints are plotted in n dimensional feature space.
if there are 3 features you can visualise plotting in 3D. I have plotted these in later code blocks.
* Interpretation of distance- for clustering various distance metrics are used. Like eucledian, manhattan, cosine(this is the cosvalue between vectors). The distance metric used will majorly affect the clustering.
* Kmeans by default uses euclidian distance. Data dictates metrics used.
If you change the feature set size the distance will change and clusters reformed or deformed.
* Use Principal component analysis for dimesionality reduction
There are functions which do this auto.
* Use feature engg to modify feature set for kmean algo. Use feture set that gives good grouping that can be interpreted.



```python
optimal_k = 0
import yfinance as yf
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
from scipy.stats import skew
from sklearn.preprocessing import StandardScaler
from tabulate import tabulate
import matplotlib.pyplot as plt

# Importing from my google drive public fething stocks listed in nifty50
#Downloaded from nse website and stored in my public folder
# There are approx 252 trading days in a year- not sure if this is amerrican or indian standard.

url = "https://drive.google.com/uc?export=download&id=1UgVkw3YibR-1WA1R-XQLHPvCVd7lRrqI"
nifty = pd.read_csv(url)
nifty50_symbols = [symbol + ".NS" for symbol in nifty["Symbol"]]

data = yf.download(nifty50_symbols, period="1y")["Adj Close"]

# Feature Engineering- the four features i have chosed for my analysis
returns = data.pct_change().dropna() # this generates a na value in the first point
volatility = returns.std() * np.sqrt(252) # this is annulised volatality
sharpe_ratio = returns.mean() / returns.std() * np.sqrt(252)
skewness = returns.skew()

# Fetch P/E ratios using yfinance
pe_ratios = []
for symbol in nifty50_symbols:
    ticker = yf.Ticker(symbol)
    info = ticker.info
    pe_ratio = info.get("trailingPE", np.nan)  # Get trailing P/E, handle missing values
    pe_ratios.append(pe_ratio)

# Create DataFrame for clustering
features = pd.DataFrame(
    {
        "Return": returns.mean() * 252, # no of trading days in a year
        "Volatility": volatility,
        "Sharpe Ratio": sharpe_ratio,
        "P/E Ratio": pe_ratios,
        "Skewness": skewness,

        # we can add or delete feature from this df to change analysis
    },
    index=data.columns,
)

# Impute missing values (with mean) else Kmeans algo will fail
#There are null values generated due to feature engg plus some likely errors in yfin
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")
features_imputed = imputer.fit_transform(features)


# Scale the features
# We need to scale features before kmean else features will not have equal influence on dist
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_imputed)  # Scale after imputation

# Elbow Method to determine optimal k
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0,n_init='auto')
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)


# Plot the elbow graph
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.show()

```

## Number of clusters
Choose number of clusters based on the plot and enter in form. Run the code


*   Tabulate the distance between sectors
*   Note the distances are scaled.


```python
optimal_k = 5 # @param {type:"integer"}
'''
kmeans? ----gives brief docu of algo with parameters and attibutes
rest get from sklearn site, imp attributes are labels and feature names.

'''


# Admanipulate clusters  based on elbow method
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
clusters = kmeans.fit_predict(scaled_features)

# Calculate distances to cluster centers
distances = kmeans.transform(scaled_features)

# Create table of cluster distances
cluster_distances = []
for i in range(optimal_k):
    cluster_distances.append(distances[clusters == i].mean(axis=0).tolist())

table_data = [["Cluster " + str(i)] + dist for i, dist in enumerate(cluster_distances)]
headers = ["Cluster"] + ["Distance to Cluster " + str(i) for i in range(optimal_k)]
print(tabulate(table_data, headers, tablefmt="fancy_grid"))
```

## List sctocks in each sector
* These stocks have been automatically classified based on the feature sets we selected for kmeans
* Each cluster will have similar stocks
* the key diversification strategy tried is that we should select stocks from different clusters to balance

```python
# Add cluster labels to the features DataFrame
features["Cluster"] = clusters
features_reset = features.reset_index().rename(columns={"index":"Ticker"})

# Create tabulate list for each cluster
cluster_data = {}
for i in range(optimal_k):
    cluster_stocks = features_reset[features_reset["Cluster"] == i]
    cluster_data[i] = cluster_stocks[["Ticker", "Return", "Volatility", "Sharpe Ratio", "P/E Ratio", "Skewness"]].values.tolist()

# Print tabulate tables
for cluster, data in cluster_data.items():
    print(f"\nCluster {cluster}:")
    print(tabulate(data, headers=["Ticker", "Return", "Volatility", "Sharpe Ratio", "P/E Ratio", "Skewness"], tablefmt="fancy_grid"))
```

## Plotting fetures in 3D
Aim of this excercise was to understand how the ndimensional feature space can be interpreted.
* We can see grouping of stock in 3d
* All nC3 combinations of feture space have been taken for plotting.

```python
import yfinance as yf
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
from scipy.stats import skew
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ... (rest of the code for fetching data, feature engineering,
#      elbow method, k-means clustering, and distance calculation remains the same)

# 3D Plotting (Modified with Legend)
feature_combinations = [
    ("Return", "Volatility", "Sharpe Ratio"),
    ("Return", "Volatility", "P/E Ratio"),
    ("Return", "Sharpe Ratio", "P/E Ratio"),
    ("Volatility", "Sharpe Ratio", "P/E Ratio"),
]

for feature1, feature2, feature3 in feature_combinations:
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111, projection="3d")

    # Create scatter plot with unique colors for each cluster
    for cluster_label in np.unique(clusters):
        cluster_points = features_reset[features_reset["Cluster"] == cluster_label]
        ax.scatter(cluster_points[feature1], cluster_points[feature2], cluster_points[feature3],
                   label=f"Cluster {cluster_label}", s=50)  # Increased marker size for visibility

    # Shortened labels for less clutter
    for j, txt in enumerate(features.index):
        short_label = txt.split(".")[0][:4]
        ax.text(features[feature1][j], features[feature2][j], features[feature3][j], short_label)

    ax.set_xlabel(feature1)
    ax.set_ylabel(feature2)
    ax.set_zlabel(feature3)
    ax.set_title(f"Cluster Distribution: {feature1}, {feature2}, {feature3}")

    # Add legend
    ax.legend()

    plt.show()
```
